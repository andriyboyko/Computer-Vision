```python
#!pip3 install opencv-python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
im=cv2.imread('lab2.jpg',  cv2.IMREAD_GRAYSCALE)
```

```python
def f(image, kernel):
    height_kernel, width_kernel = np.shape(kernel)
    image = np.pad(image, 3, mode='edge')
    shape = image.shape
    height_image, width_image = image.shape

    output = np.zeros((height_image-height_kernel, width_image-width_kernel))
    #print(height_image-height_kernel)
    for i in range(height_image-height_kernel):
        for j in range(width_image-width_kernel):
            height1, width1 = i, j
            height2, width2 = height1 + height_kernel, width1 + width_kernel
            output[i, j] = np.sum(
              image[height1:height2, width1:width2] *
              kernel, axis=(0,1))

    output = (output - output.min(axis=0)) / (output.max(axis=0) - output.min(axis=0))
    return output * 255
```

```python
#kernel=[[-1,-2,-1],[0,0,0],[-1,-2,-1]]
kernel = [[0,-1,0],[-1,4,-1],[0,-1,0]]
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1, 4, 1)
imgplot = plt.imshow(im, cmap='gray')


ax = fig.add_subplot(1, 4, 2)
imgplot = plt.imshow(f(im, kernel), cmap='gray')
```
